```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuracin global de estilo (solo fondo)
# -------------------------------------------
sns.set_style("whitegrid")

df = pd.read_csv("salary.csv", encoding="latin1", on_bad_lines="skip")
print("Primeras 5 filas:\n", df.head())

# ------------------------------------------------------------------------------
# 1) Grfico de barras (countplot) de la variable 'salary'
#    Muestra cuntas personas hay en cada categora de salario.
#    Usamos la paleta "Set2" para alternar colores ms vibrantes.
# ------------------------------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='salary', palette="Set2")
plt.title('Distribucin de salario')
plt.xlabel('Salario')
plt.ylabel('Conteo')
plt.show()

# ------------------------------------------------------------------------------
# 2) Histograma de la variable 'age'
#    Como es un histograma sin categoras en el eje X, usaremos un solo color.
#    Elegimos un tono ms intenso que "skyblue", por ejemplo "darkcyan".
# ------------------------------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='age', bins=20, edgecolor='black', color='darkcyan')
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# ------------------------------------------------------------------------------
# 3) Grfico de barras de la variable 'education' vs. la media de 'hours-per-week'
#    Usamos la paleta "Paired" para colores variados y vibrantes en cada categora.
# ------------------------------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='education', y='hours-per-week', ci=None, palette="Paired")
plt.title('Horas por semana segn nivel educativo')
plt.xlabel('Nivel Educativo')
plt.ylabel('Horas por Semana (promedio)')
plt.xticks(rotation=45)  # Rotamos las etiquetas para mayor legibilidad
plt.show()

# ------------------------------------------------------------------------------
# 4) Boxplot de 'hours-per-week' separado por 'salary'
#    Para ver cmo vara la cantidad de horas trabajadas segn la categora salarial.
#    Usamos "Set3" para variar un poco la paleta.
# ------------------------------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='salary', y='hours-per-week', palette="Set3")
plt.title('Distribucin de horas por semana segn Salary')
plt.xlabel('Salario')
plt.ylabel('Horas por Semana')
plt.show()

# ------------------------------------------------------------------------------
# 5) Grfico de barras (countplot) de 'occupation' separado por 'salary'
#    Para ver cmo se distribuyen las ocupaciones segn la categora salarial.
#    Usamos la paleta "Dark2" para alternar colores ms oscuros.
# ------------------------------------------------------------------------------
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='occupation', hue='salary', palette="Dark2")
plt.title('Ocupaciones segn salario')
plt.xlabel('Ocupacin')
plt.ylabel('Conteo')
plt.xticks(rotation=45)  # Rotamos las etiquetas para mayor legibilidad
plt.legend(title='Salary')
plt.show()
```
